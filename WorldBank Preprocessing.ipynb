```python
import pandas as pd
```

```python
df = pd.read_csv('wb_data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
transaction_columns = [
    'transaction_transaction_type_code',
    'transaction_transaction_date_iso_date',
    'transaction_value',
    'recipient_country_code',
    'iati_identifier'
]
wb_transaction = df[transaction_columns].dropna(subset=['transaction_transaction_type_code', 'transaction_transaction_date_iso_date', 'transaction_value', 'recipient_country_code', 'iati_identifier'])
wb_transaction
```

```python
wb_transaction['transaction_transaction_type_code'] = wb_transaction['transaction_transaction_type_code'].str.split(',')
```

```python
wb_transaction['transaction_transaction_date_iso_date'] = wb_transaction['transaction_transaction_date_iso_date'].str.split(',')
```

```python
wb_transaction['transaction_value'] = wb_transaction['transaction_value'].str.split(',')
```

```python
wb_transaction_exploded = wb_transaction.explode(['transaction_transaction_type_code', 'transaction_transaction_date_iso_date','transaction_value',])
wb_transaction_exploded
```

```python
wb_transaction_exploded.info()
```

```python
sector_columns = ['sector_vocabulary', 'sector_code', 'sector_percentage', 'iati_identifier']
sector_columns
wb_sector = df[sector_columns].dropna(subset=['sector_vocabulary', 'sector_code', 'sector_percentage', 'iati_identifier'])
wb_sector
```

```python
wb_sector['sector_vocabulary'] = wb_sector['sector_vocabulary'].str.split(',')
```

```python
wb_sector['sector_percentage'] = wb_sector['sector_percentage'].str.split(',')
```

```python
wb_sector['sector_code'] = wb_sector['sector_code'].str.split(',')
```

```python
wb_sector_exploded = wb_sector.explode(['sector_vocabulary', 'sector_code', 'sector_percentage'])
```

```python
wb_sector_exploded
```

```python
wb_transaction_exploded['transaction_transaction_date_iso_date'] = pd.to_datetime(
    wb_transaction_exploded['transaction_transaction_date_iso_date'])
wb_transaction_exploded['year'] = wb_transaction_exploded['transaction_transaction_date_iso_date'].dt.year
wb_transaction_exploded
```

```python
wb_transaction_exploded = wb_transaction_exploded[(wb_transaction_exploded["transaction_transaction_type_code"] == '3')]
```

```python
wb_transaction_exploded
```

```python
wb_transaction_exploded['transaction_value'] = pd.to_numeric(wb_transaction_exploded['transaction_value'])
```

```python
wb_transaction_exploded = wb_transaction_exploded[wb_transaction_exploded['year'] >= 2021]
wb_transaction_exploded
```

```python
wb_aggregated_transactions = wb_transaction_exploded.groupby(['year', 'iati_identifier', 'recipient_country_code'], as_index=False).agg({'transaction_value': 'sum'})
```

```python
wb_aggregated_transactions
```

```python
wb_join = pd.merge(wb_sector_exploded, wb_aggregated_transactions, on='iati_identifier', how='inner')
```

```python
wb_join
```

```python
wb_join
```

```python
wb_join['sector_percentage'] = wb_join['sector_percentage'].astype(float)
```

```python
wb_join.info()
```

```python
wb_join['allocated_amount'] = wb_join['transaction_value'] * (wb_join['sector_percentage'] / 100)
```

```python
wb_join
```
